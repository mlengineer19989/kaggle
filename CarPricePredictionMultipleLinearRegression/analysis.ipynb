# データの前処理

カテゴリ変数をダミー変数に変換

```python
df_cat_dummy = pd.get_dummies(df_cat)
```

```python
df_cat_dummy
```

量的変数の相関係数確認

```python
plt.figure(figsize=(10, 10))
sns.heatmap(abs(df_num.corr()), cmap="BuGn", annot=True)
```

明らかにpriceと相関がない変数（carheight, stroke, compressionratio, peakrpm）は特徴量から削除する。

```python
corr_th = 0.6
deleted_label = abs(df_num.corr())[abs(df_num.corr())["price"] < corr_th].index
df_num_dropped = df_num.drop(deleted_label, axis=1)
```

```python
df_num_dropped
```

特徴量の中で相関の強いものを確認する。

```python
abs(df_num_dropped.corr()) > 0.7
```

- wheelbase, carlength, carwidth, curbweightの間には相関がある。
- carwidth, curbweight, enginesize, horsepowerにも相関がある。enginesizeが決定することにより、carwidth, curbweight, horsepowerが決定するため、特徴量としてenginesizeを選択する。

```python
selected_chara = ['carlength', 'enginesize', 'citympg', 'price']
df_num_dropped = df_num_dropped[selected_chara]
```

```python
df_num_dropped.corr().abs() > 0.7
```

数値データの正規化を行う。

```python
df_num_dropped
```

特徴量が負の値とならないようにMin-Maxスケーリングを実施

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df_num_dropped)
df_num_dropped = pd.DataFrame(data=scaler.transform(df_num_dropped), columns=df_num_dropped.columns)
```

```python
df_num_dropped
```

数値データとカテゴリデータを結合

```python
#df_preprocessed = pd.concat([df_num_dropped, df_cat_dummy], axis=1)
df_preprocessed = df_num_dropped
```

train-test分割を行う

```python
df_y = df_preprocessed.pop("price")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_preprocessed, df_y, test_size=0.3)
```

回帰分析を行う

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
import numpy as np
x = np.arange(0, 1, 0.1)
y = x

plt.figure()
plt.scatter(y_train, model.predict(X_train))
```
